```python
import pandas as pd

def clean_options(df):
    """
    Clean NIFTY Options data:
    - Fill missing values
    - Convert timestamps
    - Calculate ATM strike
    """
    df.fillna(method='ffill', inplace=True)
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df['atm_strike'] = (df['strike'] - df['spot']).abs().idxmin()
    return df
```
